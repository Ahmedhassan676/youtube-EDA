## **Preparing the Data set**
> **1. Dropping the columns we won't need for our analysis**

```python
us_data.drop(['tags','ratings_disabled','comments_disabled','thumbnail_link','video_error_or_removed','description'], axis=1, inplace=True)
us_data.head()
```


> **2. Converting columns "Trending_date" and "publish_time" to date object**


```python
us_data.loc[:,'publish_time']=pd.to_datetime(us_data['publish_time'])
us_data['publish_time'].head()
```

```python
us_data.loc[:,'trending_date']=pd.to_datetime(us_data['trending_date'], format='%y.%d.%m')
us_data['trending_date'].head()
```

> **3. Obtaining the category from its category id using the category_id colum**



```python
categories={2 : 'Autos & Vehicles',
1 : 'Film & Animation',
10 : 'Music',
15 : 'Pets & Animals',
17 : 'Sports',
18 : 'Short Movies',
19 : 'Travel & Events',
20 : 'Gaming',
21 : 'Videoblogging',
22 : 'People & Blogs',
23 : 'Comedy',
24 : 'Entertainment',
25 : 'News & Politics',
26 : 'Howto & Style',
27 : 'Education',
28 : 'Science & Technology',
29 : 'Nonprofits & Activism',
30 : 'Movies',
31 : 'Anime/Animation',
32 : 'Action/Adventure',
33 : 'Classics',
34 : 'Comedy',
35 : 'Documentary',
36 : 'Drama',
37 : 'Family',
38 : 'Foreign',
39 : 'Horror',
40 : 'Sci-Fi/Fantasy',
41 : 'Thriller',
42 : 'Shorts',
43 : 'Shows',
44 : 'Trailers'}
```

```python
category_list=[]
for i in us_data['category_id']:
    category_list.append(categories[i])
us_data['category']=category_list
us_data.head()
```

**We drop the category_id now and use a copy of our prepared dataset**

```python
us_data_prep=us_data.iloc[:,[0,2,10,3,5,1,6,7,8,9]]
us_data_prep.head()
```



> **4. Dealing with repeated rows for each video**

**we'll be removing the repeated rows for each video and adding a new column representing the difference between the last day in trend and the 1st day the video went trending*

```python
#Calculating trending days for each video
min_values=us_data_prep.groupby('video_id').min().reset_index()
max_values=us_data_prep.groupby('video_id').max().reset_index()
video_dict={}
for i in min_values['video_id']:
    min_date=min_values[min_values['video_id']==i]['trending_date'].values
    max_date=max_values[max_values['video_id']==i]['trending_date'].values
    video_dict[i]=(pd.to_datetime(max_date)-pd.to_datetime(min_date)).days[0]
video_dict
```

```python
us_data_prep.sort_values(by='trending_date', inplace=True)
```

```python
us_data_prep.drop_duplicates(subset=['video_id'],keep='last', inplace=True)
us_data_prep.info()
```

```python
us_data_prep=us_data_prep.reset_index(drop=True)
us_data_prep
```

**Checking the Call of duty video, the last trending date should be 14/06/218**

```python
us_data_prep[us_data_prep['video_id']=='ooyjaVdt-jA']
```

**this verfies that only the last day in trend is in the data se as well as our 6351 unique video_id , Now we neew to add the trending days column to our dataset**

```python
us_data_prep['days_in_trend']=np.arange(len(us_data_prep))
```

```python
for i in video_dict:
    us_data_prep['days_in_trend']=np.where(us_data_prep['video_id']==i,video_dict[i]+1, us_data_prep['days_in_trend'])
us_data_prep.sample(2)
```

```python
us_data_prep[us_data_prep['days_in_trend']==0].count()
```

### **Analyzing our dataset**



> **1. Quick look at categories**

**We Look now at the data in hand and try to figure out the best trending categories by looking and the views, likes and dislikes count.**

```python
cat_sum=us_data_prep.groupby('category')['views','likes','dislikes','comment_count'].sum().sort_values(by='views', ascending=False)
cat_sum
```

```python
plt.figure(figsize=(9,6))
plt.bar(x=cat_sum.index.values, height=cat_sum['views'],color='midnightblue')
plt.xticks(rotation=90)
plt.title('Category views')
plt.show()
```

```python
plt.figure(figsize=(9,6))
plt.bar(x=cat_sum.index.values, height=cat_sum['comment_count'],color='midnightblue')
plt.xticks(rotation=90)
plt.title('Category comments')
plt.show()
```

**We can Look on people's engagment percentage based on how many comments there were per views and likes per views and dislikes per likes percentages**

```python
comment_percentage=np.array(cat_sum['comment_count'])/np.array(cat_sum['views'])
```

```python
plt.figure(figsize=(9,6))
plt.bar(x=cat_sum.index.values, height=comment_percentage*100,color='midnightblue')
plt.xticks(rotation=90)
plt.title('Category comments percent of views')
plt.show()
```

```python
likes_percentage=np.array(cat_sum['likes'])/np.array(cat_sum['views'])
plt.figure(figsize=(9,6))
plt.bar(x=cat_sum.index.values, height=likes_percentage*100,color='midnightblue')
plt.xticks(rotation=90)
plt.title('Category likes percent of views')
plt.show()
```

```python
dislikes_percentage=np.array(cat_sum['dislikes'])/np.array(cat_sum['likes'])
plt.figure(figsize=(9,6))
plt.bar(x=cat_sum.index.values, height=dislikes_percentage*100,color='midnightblue')
plt.xticks(rotation=90)
plt.title('Category dislikes percent of likes')
plt.show()
```

we notice that people engaged heavily in the nonprofit and activism category almost five times the gaming category (the 2nd in engagment), seems like the Music videos even if it had the most views their viewers are one of the least active.
also the News and politics category had the highest dislike percentage which may the represent this period of time (november 2017 to june 2018) political polarization that was utmost
